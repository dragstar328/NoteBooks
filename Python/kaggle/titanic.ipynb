```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import csv
from sklearn.ensemble import RandomForestClassifier
%matplotlib inline

# machine learning
from sklearn.neural_network import MLPClassifier
```

```python
train_df = pd.read_csv('../../../Data/kaggle/titanic/train.csv')
train_df.head()
```

```python
test_df = pd.read_csv('../../../Data/kaggle/titanic/test.csv')
test_df.head()
```

```python
# Gender列から　1:male,2f:emaleで新しい列を作る
train_df["Gender"] = train_df["Sex"].map({"female":0, "male":1}).astype(int)
train_df.head()

test_df["Gender"] = test_df["Sex"].map({"female":0, "male":1}).astype(int)
test_df.head()
```

```python
# Servived列から　1:Servived,0:deadで新しい列を作る
#train_df["Survived_"] = train_df["Survived"].map({1:"Servived", 0:"dead"})
#train_df.head()
```

```python
# Ageの欠損値を平均値で置換
train_df["Age"].fillna(train_df["Age"].mean().astype(int), inplace=True)
test_df["Age"].fillna(test_df["Age"].mean().astype(int), inplace=True)
```

```python
# NNet分類機をインスタンス化
clf = MLPClassifier(solver="lbfgs", alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
```

```python
# 不要な列をdrop
#train_df = train_df.drop(["Name", "Ticket", "Sex", "SibSp", "Parch", "Fare", "Cabin", "Embarked", "PassengerId", "Survived"], axis=1)
train_df = train_df.drop(["Name", "Ticket", "Sex", "SibSp", "Parch", "Fare", "Cabin", "Embarked", "PassengerId"], axis=1)
train_df.head()
```

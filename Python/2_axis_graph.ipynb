```python
import pandas as pd
import numpy
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.DataFrame(
    {
      'temp' : [27.1, 26.4, 25.8, 25.5, 24.9, 24.6, 24.8, 26, 27.6, 29.4, 30.7, 32.1, 33.4, 34.6, 34.6, 33.4, 32.1, 31, 29.5, 28.2, 27.5, 27.2, 26.5, 25.6],
      'hum' : [84, 86, 88, 89, 91, 93, 92, 88, 82, 76, 68, 60, 56, 54, 54, 56, 58, 62, 68, 70, 72, 76, 78, 82],
      'point' : [33.31522, 34.17698, 33.6724, 33.7098, 33.45111, 33.3599, 33.43516, 33.5842, 33.71446, 34.06428, 34.49427, 34.97707, 35.1736, 35.2104, 35.12026, 35.22698, 34.64425, 34.2211, 34.38615, 34.321, 34.5087, 34.19452, 34.09405, 33.50762]
    })
df.head()
```

```python


fig, ax1 = plt.subplots()

s = df.index
ax1.bar(s ,df['point'], color = 'g')
ax1.set_ylim([0,50])
ax1.axhline(28, color="c", ls="dashed", label='warn')
ax1.axhline(32, color="m", ls="dashed", label='alert')
ax1.set_ylabel("POINT")
ax1.set_xlabel("HOUR")

ax2=ax1.twinx()
ax2.plot(s,df['temp'], color="r", label = 'temp')
ax2.plot(s,df['hum'], color="b", label = 'hum')
ax2.set_ylabel("temp / hum")

plt.legend()
plt.show()

```

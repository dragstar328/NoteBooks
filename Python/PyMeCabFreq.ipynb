```python
import MeCab
import pandas as pd
```

```python
mt = MeCab.Tagger('')
```

```python
f = open("../../Data/trump.txt", encoding='utf8')
mdf = pd.DataFrame(columns=['Term', 'Info1', 'Info2', 'Freq'])
line = f.readline()
dic = {}
while line:
    node = mt.parseToNode(line)
    while node:
        arr = node.feature.split(',')     
        if arr[0].startswith("BOS/EOS"):
            pass
        else:
            s = dic.get(arr[6], pd.Series([arr[6], arr[0], arr[1], 0], index=mdf.columns))
            s['Freq'] += 1
            dic[arr[6]] = s
        node = node.next
    line = f.readline()

for v in dic.values():
    mdf = mdf.append(v, ignore_index=True)

f.close()
mdf.head()
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from sympy import * 
font = {"family": "TakaoGothic"}
matplotlib.rc('font', **font)
init_printing()
```

```python
x, y = symbols('x, y', real=True)
```

```python
fs = [
    (x ** 4 - 3 * x ** 2 + 1, -1),
    (x ** 3 + x - 2, 0),
    (-x ** 3 + 2 * x + 1, 2),
    (2 * x ** 3 + 5, 21),
    (5 * x ** 2 + 1, 11)
]
```

```python
for i, (f, y0) in enumerate(fs, 6):
    print(f'({i})')
    pprint(f)
    s = solve(y -f, x)
    for g in s:
        g1 = Derivative(g, y, 1)
        for h in [g1, g1.doit(), g1.doit().subs({y: y0})]:
            pprint(h)
            print()
    p = plot(
        x, f, *map(lambda s0: s0.subs({y: x}), s), show=False, legend=True
    )
    
    for j, _ in enumerate(p):
        if j == 0:
            p[j].line_color = 'red'
        elif j == 1:
            p[j].line_color = 'green'
        else:
            p[j].line_color = 'blue'
    p.show()
    print()
```
